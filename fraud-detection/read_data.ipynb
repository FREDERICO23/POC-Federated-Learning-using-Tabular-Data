### Data Preprocessing

```python
df = data.copy()
```

```python
# separating the data for analysis and data sample creation
legit = df[df.Class == 0]
fraud = df[df.Class == 1]
```

```python
print("Fraud Transactions: ",fraud.shape)
print("Legit Transactions: ", legit.shape)
```

```python
## Take a sample of atleast 90% more of the Fraud transaction to reduce biasness in the data
legit_sample = legit.sample(n=4920)

## Do for another datasets for federated learning
legit_sample1 = legit.sample(n=4000)
```

```python
# Concatenante the legit sample and the fraud to create a less biased dataset
import random

dfs = [legit_sample, fraud]
random.shuffle(dfs)
new_df = pd.concat(dfs, axis=0)

## Do for another datasets for federated learning
dfs = [legit_sample1, fraud]
random.shuffle(dfs)
df1 = pd.concat(dfs, axis=0)


new_df.shape
```

```python
df1
```

```python
new_df.groupby('Class').mean()
```

```python
X = new_df.drop(columns = 'Class',axis = 1)
y = new_df['Class']
```

### Modelling

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score 
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2, stratify = y, random_state = 23)
print(X_train.shape, X_test.shape, y_train.shape,y_test.shape)
```

```python
lr =  LogisticRegression()
lr.fit(X_train,y_train)
```

```python
preds = lr.predict(X_test)
```

```python
pred = accuracy_score(preds, y_test)
pred
```

### Using SMOTE to solve data imbalance
